```python
# Rellenamos los valores nulos de la columna 'neighborhood_overview' con 'S/D'
df1['neighborhood_overview'] = df1['neighborhood_overview'].fillna('S/D')
```

```python
# Rellenamos los valores nulos de la columna 'host_about' con 'S/D'
df1['host_location'] = df1['host_location'].fillna('S/D')
```

```python
# Rellenamos los valores nulos de la columna 'host_about' con 'S/D'
df1['host_about'] = df1['host_about'].fillna('S/D')
```

```python
# Rellenamos los valores nulos de la columna 'host_response_time' con 'S/D'
df1['host_response_time'] = df1['host_response_time'].fillna('S/D')
```

```python
# Rellenamos los valores nulos de la columna 'host_response_rate' con 'S/D'
df1['host_response_rate'] = df1['host_response_rate'].fillna('S/D')
```

```python
# Rellenamos los valores nulos de la columna 'host_acceptance_rate' con 'S/D'
df1['host_acceptance_rate'] = df1['host_acceptance_rate'].fillna('S/D')
```

```python
# Rellenamos los valores nulos de la columna 'host_is_superhost' con 'f' (valor booleano)
df1['host_is_superhost'] = df1['host_is_superhost'].fillna("f")
```

```python
# Rellenamos los valores nulos de la columna 'bathrooms_text' con '1 bath'
df1['bathrooms_text'] = df1['bathrooms_text'].fillna('1 bath')
```

```python
# Rellenamos los valores nulos de la columna 'bedrooms' con '1'
df1['bedrooms'] = df1['bedrooms'].fillna(1)
```

```python
# Rellenamos los valores nulos de la columna 'beds' con '1'
df1['beds'] = df1['beds'].fillna(1)
```

```python
# Para la columna price, primero convertimos los valores a num√©ricos y despu√©s interpolamos los valores nulos
df1['price'] = df1['price'].str.replace('$', '')
df1['price'] = df1['price'].str.replace(',', '')
df1['price'] = pd.to_numeric(df1['price'])
df1['price'] = df1['price'].interpolate()
```

```python
# Rellenamos los valores nulos de la columna 'has_availability' con 'f' (valor booleano)
df1['has_availability'] = df1['has_availability'].fillna('f')
```

```python
# Rellenamos las columnas 'first_review' y 'last_review' con 'S/D'
df1['first_review'] = df1['first_review'].fillna('S/D')
df1['last_review'] = df1['last_review'].fillna('S/D')
```

```python
# Rellenamos los valores nulos de las columnas de review_scores_... con 0
df1['review_scores_rating'] = df1['review_scores_rating'].fillna(0)
df1['review_scores_accuracy'] = df1['review_scores_accuracy'].fillna(0)
df1['review_scores_cleanliness'] = df1['review_scores_cleanliness'].fillna(0)
df1['review_scores_checkin'] = df1['review_scores_checkin'].fillna(0)
df1['review_scores_communication'] = df1['review_scores_communication'].fillna(0)
df1['review_scores_location'] = df1['review_scores_location'].fillna(0)
df1['review_scores_value'] = df1['review_scores_value'].fillna(0)
```

```python
# Rellenamos los valores nulos de la columna 'reviews_per_month' con la interpolaci√≥n de los valores
df1['reviews_per_month'] = df1['reviews_per_month'].interpolate()
# Quitamos el √∫nico valor nulo restante
df1 = df1.dropna()
```

```python
# Realizamos la comprobaci√≥n de que no hay valores nulos en el dataset
df1.isnull().sum()
```

### Tratamiento de Valores At√≠picos

```python
df1.info()
```

```python
# separamos las variables cualitativas y cuantitativas
cualitativas = df1.select_dtypes(include=['object'])
cuantitativas = df1.select_dtypes(include=['float64', 'int64'])
```

```python
# El dataset cuenta con las variables 'id' y 'host_id' que no son relevantes para el an√°lisis
# por lo que las movemos al principio del dataset junto con las variables cualitativas
cualitativas = pd.concat([df1['id'], df1['host_id'], cualitativas], axis=1)
cuantitativas = cuantitativas.drop(['id', 'host_id', 'latitude', 'longitude'], axis=1)
```

```python
cualitativas.columns
```

```python
cuantitativas.columns
```

```python
# Realizamos un diagrama de cajas para las variables cuantitativas
fig = plt.figure(figsize=(30, 30))
cuantitativas.plot(kind='box', vert=False)
plt.title('Valores At√≠picos de las Variables Cuantitativas')
plt.show()
```

```python
# realizamos el diagrama de caja de cada columna de forma individual
for column in cuantitativas.columns:
    plt.figure(figsize=(10, 3))
    cuantitativas.boxplot(column=column, vert=False)
    plt.title(f'Valores At√≠picos de la Variable: {column}')
    plt.show()
```

### M√©todo "Desviaci√≥n Estandar" Para Detectar Valores At√≠picos

```python
# M√©todo aplicando la desviaci√≥n est√°ndar para encontrar los valores extremos
# Se considera que un valor es extremo si est√° a m√°s de 3 desviaciones est√°ndar de la media
y=cuantitativas
Limite_Superior=y.mean() + 3*y.std()
Limite_Inferior=y.mean() - 3*y.std()
print("------------- Limite Superior -------------- \n", Limite_Superior)
print("------------- Limite Inferior ------------- \n", Limite_Inferior)
```

```python
# obtenemos los datos y los outliers se convierten en NaN
df2 = cuantitativas[(y <= Limite_Superior) & (y >= Limite_Inferior)]
df2
```

```python
# corroboramos valores nulos
df2.isnull().sum()
```

```python
# realizamos una copia del dataframe y remplazamos valores nulos por la media
df_clean = df2.copy()
df_clean.fillna(round(df2.mean(), 1), inplace=True)
df_clean.isnull().sum()
```

```python
# realizamos el diagrama de caja
fig = plt.figure(figsize=(30, 30))
df_clean.plot(kind='box', vert=False)
plt.title('Valores At√≠picos de las Variables Cuantitativas')
plt.show()
```

M√©todo "Cuantiles" para Detectar Valores At√≠picos

```python
# m√©todo aplicando rango intercuart√≠lico para detectar outliers. Encuentro cuartiles 0.25 y 0.75

y = cuantitativas

percentil25 = y.quantile(0.25) # Q1
percentil75 = y.quantile(0.75) # Q3
IQR = percentil75 - percentil25

Limite_Superior_IQR = percentil75 + 1.5 * IQR
Limite_Inferior_IQR = percentil25 - 1.5 * IQR

print("------------ Limite Superior -------------  \n", Limite_Superior_IQR)
print("------------ Limite Inferior -------------  \n", Limite_Inferior_IQR)
```

```python
# obtenemos los datos y los outliers se convierten en NaN
df3 = cuantitativas[(y <= Limite_Superior_IQR) & (y >= Limite_Inferior_IQR)]
df3.isnull().sum()
```

```python
# realizamos una copia del dataframe y remplazamos valores at√≠picos por la media
df_IQR = df3.copy()
df_IQR.fillna(round(df3.mean(), 1), inplace=True)
df_IQR
```

```python
fig = plt.figure(figsize=(30, 30))
df_IQR.plot(kind='box', vert=False)
plt.title('Valores At√≠picos de las Variables Cuantitativas')
plt.show()
```

```python
# unimos las variables cualitativas con las cuantitativas
dataset_limpio = pd.concat([cualitativas, df_IQR], axis=1)
dataset_limpio.isnull().sum()
```
